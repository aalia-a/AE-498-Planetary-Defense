```python
from sympy import *
import numpy as np

def R1(theta):
    return np.array([
        [1, 0, 0],
        [0, np.cos(theta), -np.sin(theta)],
        [0, np.sin(theta),  np.cos(theta)]
    ])

def R3(theta):
    return np.array([
        [ np.cos(theta), -np.sin(theta), 0],
        [ np.sin(theta),  np.cos(theta), 0],
        [0, 0, 1]
    ])

def r_vect(a, e, i, arg_of_peri, RAAN, f):
    p = a*(1 - e**2)
    r = p / (1 + e*np.cos(f))
    #arg_of_peri = np.deg2rad(arg_of_peri) #angles given in degrees (convert to radians)
    #RAAN = np.deg2rad(RAAN)
    #i = np.deg2rad(i)
    r_pqw = r*np.array([[np.cos(f)], [np.sin(f)], [0]])
    Q = R3(RAAN) @ R1(i) @ R3(arg_of_peri)
    return Q @ r_pqw


def distance_squared(f1, f2, orbit1, orbit2):
  r1 = r_vect(*orbit1, f1)
  r2 = r_vect(*orbit2, f2)
  d_squared = (np.linalg.norm(r1 - r2))**2
  return d_squared


def MOID(orbit1, orbit2):
  N = 800
  f = np.linspace(0, 2*np.pi, N)
  f1, f2 = None, None
  d_min = np.inf
  for f1 in f:
    for f2 in f:
      d_squared = distance_squared(f1, f2, orbit1, orbit2)
      if d_squared < d_min:
          d_min = d_squared
          f_min = (f1, f2)
  return f_min, np.sqrt(d_min)

def km_to_AU(distance):
  return distance / 149597870.691 # from https://www.unitconverters.net/length/astronomical-unit-to-kilometer.htm
```

```python
# question 1
# The Earth and asteroid 99942 Apophis

a_earth, e_earth, i_earth, arg_of_peri_earth ,RAAN_earth = 1.47650607e+08, 9.1669995e-03, 4.2422693e-03, 6.04375107e+01, 1.4760836e+01
a_99942, e_99942, i_99942, arg_of_peri_99942 ,RAAN_99942 = 1.3793939e+08,  1.9097084e-01, 3.3356539e+00,  1.2919949e+02, 2.0381969e+02

earth_orbit = (a_earth, e_earth, np.deg2rad(i_earth), np.deg2rad(arg_of_peri_earth), np.deg2rad(RAAN_earth))
apophis_orbit = (a_99942, e_99942, np.deg2rad(i_99942), np.deg2rad(arg_of_peri_99942), np.deg2rad(RAAN_99942))

f_min1, moid_km1 = MOID(earth_orbit, apophis_orbit)
moid_au1 = km_to_AU(moid_km1)

print(f"f1 = {f_min1[0]} rad, f2 = {f_min1[1]} rad")
print(f"MOID: {moid_km1} km = {moid_au1} AU")
```
